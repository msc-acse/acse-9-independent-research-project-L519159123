```python
%matplotlib inline
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
dir = './results/seismic_cyclegan_res9_crop_900_2/test_latest/files/'
filename = 'ep0001189_1_fake_B.txt'
filename2 = 'ep0001189_1_real_B.txt'

filename3 = 'ep0001189_1_real_A.txt'
filename4 = 'ep0001189_1_rec_A.txt'
with open(os.path.join(dir, filename), 'rb') as fh:
    data = np.genfromtxt(fh, delimiter=',', dtype='float64').astype('float32')
    
with open(os.path.join(dir, filename2), 'rb') as fh:
    data2 = np.genfromtxt(fh, delimiter=',', dtype='float64').astype('float32')
    
with open(os.path.join(dir, filename3), 'rb') as fh:
    data3 = np.genfromtxt(fh, delimiter=',', dtype='float64').astype('float32')
    
with open(os.path.join(dir, filename4), 'rb') as fh:
    data4 = np.genfromtxt(fh, delimiter=',', dtype='float64').astype('float32') 

fig = plt.figure(figsize=(8, 8))
ax1 = fig.add_subplot(2, 2, 1)
ax1 = plt.gca()
ax2 = fig.add_subplot(2, 2, 2)
ax2 = plt.gca()
ax3 = fig.add_subplot(2, 2, 3)
ax3 = plt.gca()
ax4 = fig.add_subplot(2, 2, 4)
ax4 = plt.gca()

ax1.set_xlabel('receiver index', fontsize=12)
ax1.set_ylabel('time (s)', fontsize=12)
ax1.xaxis.tick_top()
ax1.xaxis.set_label_position('top')
ax1.text(10, 90, 'real_P', fontsize=12)
ax1.set_xticklabels([0, 48, 96, 144, 192, 240])
ax1.set_yticklabels([7.6, 6.08, 4.56, 3.04, 1.52, 0])
ax1.imshow(data3, origin='upper', vmin=-0.1, vmax=0.1, cmap=plt.cm.seismic, extent=[0, 100, 0, 100])

ax2.set_xlabel('receiver index', fontsize=12)
ax2.xaxis.tick_top()
ax2.xaxis.set_label_position('top')
ax2.text(10, 90, 'rec_P', fontsize=12)
ax2.set_xticklabels([0, 48, 96, 144, 192, 240])
ax2.set_yticklabels([])
ax2.tick_params(top='on', bottom='off',left='off')
ax2.imshow(data4, vmin=-0.1, vmax=0.1, cmap=plt.cm.seismic, extent=[0, 100, 0, 100])

ax3.set_xlabel('receiver index', fontsize=12)
ax3.set_ylabel('time (s)', fontsize=12)
ax3.xaxis.tick_top()
ax3.xaxis.set_label_position('top')
ax3.text(10, 90, 'fake_Z', fontsize=12)
ax3.set_xticklabels([0, 48, 96, 144, 192, 240])
ax3.set_yticklabels([7.6, 6.08, 4.56, 3.04, 1.52, 0])
ax3.imshow(data, origin='upper', vmin=-0.1, vmax=0.1, cmap=plt.cm.seismic, extent=[0, 100, 0, 100])

ax4.set_xlabel('receiver index', fontsize=12)
ax4.xaxis.tick_top()
ax4.xaxis.set_label_position('top')
ax4.text(10, 90, 'real_Z', fontsize=12)
ax4.set_xticklabels([0, 48, 96, 144, 192, 240])
ax4.set_yticklabels([])
ax4.tick_params(top='on', bottom='off',left='off')
ax4.imshow(data2, vmin=-0.1, vmax=0.1, cmap=plt.cm.seismic, extent=[0, 100, 0, 100])

fig.tight_layout()

directory = './results/seismic_cyclegan_res9_crop_900_2/test_latest/'
plt.savefig(os.path.join(directory, 'temp.png'))
```
